```python
import numpy as np
import matplotlib.pyplot as plt

# Define Parameter Values
alpha = 0.2
beta = 0.8
gamma = 0.1 


# Define Step
h = 0.01

# Define Functions
def func1(y):
    return -alpha * y

def func2(x, y):
    return beta * x - gamma * y**2

# Initial Conditions
x0 = 2
y0 = 0

# Define Time Span
t0 = 0
tf = 60

# Runge - Kutta 4th order 
def RungeKutta4thOrder(func1, func2, x0, y0, t0, tf, h):
    n = int((tf - t0) / h)
    
    t = np.linspace(t0, tf, n + 1)
    x = np.zeros(n + 1)
    y = np.zeros(n + 1)

    x[0] = x0
    y[0] = y0

    for i in range(n):
        k0_1 = h * func1(y[i])
        k1_1 = h * func1(y[i] + 1/2 * k0_1)
        k2_1 = h * func1(y[i] + 1/2 * k1_1)
        k3_1 = h * func1(y[i] + k2_1)
        x[i + 1] = x[i]+ 1/6 * (k0_1 + 2 * k1_1 + 2 * k2_1 + k3_1)


        k0_2 = h * func2(x[i], y[i])
        k1_2 = h * func2(x[i] + 1/2 * h, y[i] + 1/2 * k0_2)
        k2_2 = h * func2(x[i] + 1/2 * h, y[i] + 1/2 * k1_2)
        k3_2 = h * func2(x[i] + h, y[i] + k2_2)
        y[i + 1] = y[i] + 1/6 * (k0_2 + 2 * k1_2 + 2 * k2_2 + k3_2)

    return x, y, t


# Solve the differential equations
x, y, t = RungeKutta4thOrder(func1, func2, x0, y0, t0, tf, h)


# Plot the results
plt.plot(t, x, label='x(t)')
plt.plot(t, y, label='y(t)')
plt.xlabel('Time t')
plt.ylabel('Values of x and y')
plt.legend()
plt.title('Runge-Kutta 4th Order Solution')
plt.grid(True)
plt.show()

```
